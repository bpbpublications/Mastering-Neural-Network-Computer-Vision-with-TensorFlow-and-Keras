```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
 return 1/(1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x, y)
plt.show()
```

```python
def relu(x):
    return max(0.0, x)

x = np.linspace(-10, 10, 100)
y = [relu(a) for a in x]
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.plot(x, y)
plt.show()
```

```python
def leaky_relu(x):
  if x>0 :
    return x
  else :
    return 0.1*x

x = np.linspace(-10, 10, 100)
y = [leaky_relu(a) for a in x]
plt.xlabel("x")
plt.ylabel("leaky_relu(x)")
plt.plot(x, y)
plt.show()
```

```python
a = 1.0
def elu(x):
    if x > 0:
        return x
    return a*(np.exp(x) - 1)

x = np.linspace(-10, 10, 100)
y = [elu(i) for i in x]
plt.xlabel("x")
plt.ylabel("elu(x)")
plt.plot(x, y)
plt.show()
```

```python
import math

x = np.linspace(-10, 10, 100)
y = [math.tanh(a) for a in x]
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.plot(x, y)
plt.show()
```

```python
import numpy as np

# actual values of the properties
y = [90,100,500,235,99, 1000000]
# predicted values
y_hat_1 = [64.5,199.0,456.5,283,45, 1000000]
y_hat_2 = [64.5,199.0,456.5,283,45, 100000]

print("1. MSE")
mse1 = np.square(np.subtract(y,y_hat_1)).mean()
print("mse_1 = ", int(mse1))
mse2 = np.square(np.subtract(y,y_hat_2)).mean()
print("mse_2 = ", int(mse2))

print("\n2. MAE")
mae1 = np.abs(np.subtract(y,y_hat_1)).mean()
print("mae_1 = ", int(mae1))
mae2 = np.abs(np.subtract(y,y_hat_2)).mean()
print("mae_2 = ", int(mae2))
```
