```python
import numpy as np
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

#Normalize the input
x_train, x_test = x_train.astype('float32')/255., x_test.astype('float32')/255.

print("Training dimensions before reshaping: ", x_train.shape)
# Reshape the input
x_train, x_test = x_train.reshape(x_train.shape[0], -1), x_test.reshape(x_test.shape[0], -1)

print("Training dimensions after reshaping: ", x_train.shape)

```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

n = 5
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
for i in range(n*n):
  index = np.random.randint(0, 59999)
  figure[(i//n)*digit_size:((i//n)+1)*digit_size, (i%n)*digit_size:((i%n)+1)*digit_size ] = x_train[index]
plt.figure(figsize=(5, 5))
plt.imshow(figure, cmap='Greys_r')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

#Normalize the input
n = 5
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))

for i in range(n*n):
  index = np.random.randint(0, 59999)
  figure[(i//n)*digit_size:((i//n)+1)*digit_size,
   (i%n)*digit_size:((i%n)+1)*digit_size ] = x_train[index]

plt.figure(figsize=(5, 5))
plt.imshow(figure, cmap='Greys_r')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
#Normalize the input
n = 5
digit_size = 32
figure = np.zeros((digit_size * n, digit_size * n, 3)).astype(int)

for i in range(n*n):
  index = np.random.randint(0, 49999)
  figure[(i//n)*digit_size:((i//n)+1)*digit_size, (i%n)*digit_size:((i%n)+1)*digit_size, 0: 3] = x_train[index]
plt.figure(figsize=(5, 5))
plt.imshow(figure)
plt.show()
```
