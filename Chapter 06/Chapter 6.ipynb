```python
import json
from PIL import Image, ImageDraw

with open("/content/via_project_10Aug2023_9h3m_json.json") as file:
  contents = json.load(file)

labels_dict = {"background": 0,
               "road": 100,
               "car": 255}

for image_data in contents.values():
  input = "/content/images/" + image_data["filename"]
  width, height = Image.open(input).size
  image = Image.new(mode = "RGB", size=(width, height), color = labels_dict["background"])
  draw = ImageDraw.Draw(image)
  for region in image_data["regions"]:
    points = [
        point for point in zip(region["shape_attributes"]["all_points_x"],
                              region["shape_attributes"]["all_points_y"])]
    label = region["region_attributes"]["type"]
    draw.polygon((points), fill=labels_dict[label])

  out_file_type = ".png"
  annotated_img = "/content/annotations/" + image_data["filename"]
  image.save(annotated_img)
```

```python
import tensorflow as tf
iou = tf.keras.metrics.IoU(num_classes=2, target_class_ids=[0])
annotated_mask = [0, 1, 1, 0]
predicted_mask = [0, 1, 1, 1]
iou.update_state(annotated_mask, predicted_mask)
iou.result().numpy()
```
