```python
# Let's assume that we have an input of shape (784,1) and that we have labels of shape (10,1)
n_hidden_1 = 256
n_hidden_2 = 128
n_hidden_3 = 64
n_input = 784
n_classes  = 10
stddev = 0.1

class MyModel(tf.Module):
 def __init__(self, **kwargs):
   super().__init__(**kwargs)

   # Initialize the weights and biases in a random way
   self.weights = {
       'h1': tf.Variable(tf.random.normal([n_input, n_hidden_1], 				stddev=stddev)),
       'h2': tf.Variable(tf.random.normal([n_hidden_1, n_hidden_2], 				stddev=stddev)),
       'h3': tf.Variable(tf.random.normal([n_hidden_2, n_hidden_3], 				stddev=stddev)),
       'out': tf.Variable(tf.random.normal([n_hidden_3, n_classes], 				stddev=stddev))
   }
   self.biases = {
       'b1': tf.Variable(tf.random.normal([n_hidden_1])),
       'b2': tf.Variable(tf.random.normal([n_hidden_2])),
       'b3': tf.Variable(tf.random.normal([n_hidden_3])),
       'out': tf.Variable(tf.random.normal([n_classes]))
   }

 def __call__(self, x):
   layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(x, self.weights['h1']), self.biases['b1']))
   layer_2 = tf.nn.sigmoid(tf.add(tf.matmul(layer_1, self.weights['h2']), self.biases['b2']))
   layer_3 = tf.nn.sigmoid(tf.add(tf.matmul(layer_2, self.weights['h3']), self.biases['b3']))
   return (tf.matmul(layer_3, self.weights['out']) + self.biases['out'])

model = MyModel()
```

```python
hidden_1_size = 256
hidden_2_size = 128
hidden_3_size = 64
input_size = 784
n_classes  = 10
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(hidden_1_size,activation = 'sigmoid',input_shape=(input_size,)))
model.add(tf.keras.layers.Dense(hidden_2_size,activation = 'sigmoid'))
model.add(tf.keras.layers.Dense(hidden_3_size,activation = 'sigmoid'))
model.add(tf.keras.layers.Dense(n_classes))
```
